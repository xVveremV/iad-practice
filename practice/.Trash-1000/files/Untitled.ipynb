```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Завантаження датасету Titanic ---
url = 'https://drive.google.com/uc?id=1zyYJMa75UXLjmODPRnl4_UrLJ0JKmEOC'
df = pd.read_csv(url)

# --- Налаштування стилю графіків ---
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (8,5)

# --- 1. Пропуски у даних ---
plt.figure()
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title("Пропущені значення у датасеті")
plt.show()

# --- 2. Кількість виживших ---
plt.figure()
sns.countplot(x='Survived', data=df)
plt.title("Кількість виживших (0 = ні, 1 = так)")
plt.show()

# --- 3. Виживання за статтю ---
plt.figure()
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Виживання за статтю")
plt.show()

# --- 4. Виживання за класом каюти ---
plt.figure()
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Виживання за класом каюти")
plt.show()

# --- 5. Розподіл віку ---
plt.figure()
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Розподіл віку пасажирів")
plt.show()

# --- 6. Boxplot віку за класами ---
plt.figure()
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Вік пасажирів за класами")
plt.show()

# --- 7. Розподіл вартості квитків ---
plt.figure()
sns.histplot(df['Fare'], bins=40, kde=True)
plt.title("Розподіл вартості квитків (Fare)")
plt.show()

# --- 8. Виживання залежно від кількості братів/сестер/подружжя (SibSp) ---
plt.figure()
sns.countplot(x='SibSp', hue='Survived', data=df)
plt.title("Виживання за кількістю родичів (SibSp)")
plt.show()

# --- 9. Виживання залежно від кількості батьків/дітей (Parch) ---
plt.figure()
sns.countplot(x='Parch', hue='Survived', data=df)
plt.title("Виживання за кількістю дітей/батьків (Parch)")
plt.show()

# --- 10. Розподіл віку за статтю ---
plt.figure()
sns.kdeplot(data=df, x='Age', hue='Sex', fill=True)
plt.title("Розподіл віку за статтю")
plt.show()

# --- 11. Boxplot вартості квитків за класами ---
plt.figure()
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Вартість квитків за класами")
plt.show()

# --- 12. Кореляційна матриця ---
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Кореляційна матриця числових ознак")
plt.show()

# --- 13. Pairplot для кількох ознак ---
sns.pairplot(df[['Survived','Age','Fare','Pclass']], hue='Survived')
plt.suptitle("Взаємозв’язки між віком, ціною, класом і виживанням", y=1.02)
plt.show()
```
