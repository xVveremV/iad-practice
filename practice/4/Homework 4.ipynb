```python
# Тема: Custom dataset generation, Linear Regression, Non-linear Regression
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
# -------------------------
# 1. Custom dataset generation
# -------------------------
```

```python
np.random.seed(42)

N = 50
X = np.linspace(0, 5, N).reshape(-1, 1)   # feature (0..5)
y_true = 2 * X.flatten() + 1              # справжня залежність (лінійна)
noise = np.random.normal(0, 1, N)         # випадковий шум
y = y_true + noise   
```

```python
# ціль з шумом
```

```python
plt.scatter(X, y, color='red', label='Data')
plt.title("Custom Dataset")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
# -------------------------
# 2. Linear Regression
# -------------------------
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)

y_pred = lin_reg.predict(X)

plt.scatter(X, y, color='red', label='Data')
plt.plot(X, y_pred, color='blue', label='Linear Regression')
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

print("Linear Regression: y =", lin_reg.coef_[0], "* x +", lin_reg.intercept_)
```

```python
# -------------------------
# 3. Non-linear Regression (Polynomial)
# -------------------------
```

```python
# згенеруємо нелінійні дані
```

```python
X_nl = np.linspace(-3, 3, 100).reshape(-1, 1)
y_nl_true = 0.5 * X_nl.flatten()**3 - X_nl.flatten()**2 + 2 * X_nl.flatten() + 3
noise_nl = np.random.normal(0, 3, len(X_nl))
y_nl = y_nl_true + noise_nl
```

```python
# поліноміальна регресія 3-го степеня
```

```python
poly_model = make_pipeline(PolynomialFeatures(3), LinearRegression())
poly_model.fit(X_nl, y_nl)

y_nl_pred = poly_model.predict(X_nl)

plt.scatter(X_nl, y_nl, color='orange', s=10, label='Data')
plt.plot(X_nl, y_nl_pred, color='blue', label='Polynomial Regression (deg=3)')
plt.title("Non-linear Regression (Polynomial)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
