```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# --- Load Iris dataset ---
```

```python
iris = load_iris()
feature_names = iris['feature_names']

df = pd.DataFrame(data=iris['data'], columns=feature_names)
df['target'] = iris['target']

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (8,5)
```

```python
# --- 1. Гістограми всіх фіч ---
```

```python
df[feature_names].hist(bins=15, figsize=(12,8))
plt.suptitle("Розподіл значень Iris фіч")
plt.show()
```

```python
# --- 2. KDE розподіл по класах ---
```

```python
for col in feature_names:
    plt.figure()
    sns.kdeplot(data=df, x=col, hue="target", fill=True)
    plt.title(f"KDE розподіл {col} по класах")
    plt.show()
```

```python
# --- 3. Boxplots по класах ---
```

```python
for col in feature_names:
    plt.figure()
    sns.boxplot(x="target", y=col, data=df)
    plt.title(f"Boxplot {col} по класах")
    plt.show()
```

```python
# --- 4. Violin plots ---
```

```python
for col in feature_names:
    plt.figure()
    sns.violinplot(x="target", y=col, data=df, inner="quartile")
    plt.title(f"Violin plot {col} по класах")
    plt.show()
```

```python
# --- 5. Stripplots ---
```

```python
for col in feature_names:
    plt.figure()
    sns.stripplot(x="target", y=col, data=df, jitter=True, alpha=0.7)
    plt.title(f"Stripplot {col} по класах")
    plt.show()
```

```python
# --- 6. Swarmplots ---
```

```python
for col in feature_names:
    plt.figure()
    sns.swarmplot(x="target", y=col, data=df, size=4)
    plt.title(f"Swarmplot {col} по класах")
    plt.show()
```

```python
# --- 7. Кореляційна матриця ---
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Кореляційна матриця Iris фіч")
plt.show()
```

```python
# --- 8. Pairplot ---
```

```python
sns.pairplot(df, hue="target", palette="Set2", diag_kind="kde")
plt.suptitle("Pairplot Iris фіч", y=1.02)
plt.show()
```
