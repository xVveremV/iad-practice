```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as dss
```

```python
# --- Завантаження датасету Wine ---
```

```python
wine = dss.load_wine()
feature_names = wine['feature_names'][:5]

df = pd.DataFrame(
    data = wine['data'][:, :5],
    columns = feature_names
)
df['target'] = wine['target']

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (8,5)
```

```python
# --- 1. Розподіл значень перших фіч (гістограми) ---
```

```python
df[feature_names].hist(bins=20, figsize=(12, 8))
plt.suptitle("Розподіл перших 5 ознак")
plt.show()
```

```python
# --- 2. KDE розподіли для кожної фічі ---
```

```python
for col in feature_names:
    plt.figure()
    sns.kdeplot(data=df, x=col, hue="target", fill=True)
    plt.title(f"KDE розподіл {col} по класах")
    plt.show()
```

```python
# --- 3. Віолін-плоти ---
```

```python
for col in feature_names:
    plt.figure()
    sns.violinplot(x="target", y=col, data=df, inner="quartile")
    plt.title(f"Violin plot для {col}")
    plt.show()
```

```python
# --- 4. Stripplot (розкид значень) ---
```

```python
for col in feature_names:
    plt.figure()
    sns.stripplot(x="target", y=col, data=df, jitter=True, alpha=0.7)
    plt.title(f"Stripplot для {col}")
    plt.show()
```

```python
# --- 5. Swarmplot (щільний розкид) ---
```

```python
for col in feature_names:
    plt.figure()
    sns.swarmplot(x="target", y=col, data=df, size=4)
    plt.title(f"Swarmplot для {col}")
    plt.show()
```

```python
# --- 6. Кореляційна матриця (heatmap) ---
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Кореляція ознак")
plt.show()

# --- 7. Pairplot (усі пари фіч) ---
sns.pairplot(df, hue="target", palette="Set1")
plt.suptitle("Pairplot для перших 5 ознак", y=1.02)
plt.show()
```
